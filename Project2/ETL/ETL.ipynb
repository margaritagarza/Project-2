```python
import pandas as pd
from sqlalchemy import create_engine
from config import key as password
```

```python
file="food.csv"
data= pd.read_csv(file)
data
```

```python
#removal of columns that we won't use from the first csv
food_analysis_df= data[['Category', 'Description', 'Nutrient Data Bank Number','Data.Carbohydrate', 'Data.Cholesterol', 'Data.Fiber', 'Data.Kilocalories', 'Data.Protein', 'Data.Sugar Total', 'Data.Water', 'Data.Fat.Monosaturated Fat', 'Data.Fat.Polysaturated Fat', 'Data.Fat.Saturated Fat', 'Data.Fat.Total Lipid', 'Data.Major Minerals.Copper', 'Data.Major Minerals.Iron', 'Data.Major Minerals.Magnesium', 'Data.Major Minerals.Phosphorus','Data.Major Minerals.Potassium', 'Data.Major Minerals.Sodium', 'Data.Major Minerals.Zinc','Data.Vitamins.Vitamin A - RAE','Data.Vitamins.Vitamin B12', 'Data.Vitamins.Vitamin B6', 'Data.Vitamins.Vitamin C', 'Data.Vitamins.Vitamin E', 'Data.Vitamins.Vitamin K']]
food_analysis_df.head()
```

```python
nutrients=pd.DataFrame(list(food_analysis_df.columns.values))
```

```python
nutrients=nutrients[0].str.split(".", expand=True)
```

## Essential Nutrient Group

```python
#Create essential nutrient group table
essential_nutrient_group= nutrients[[1]].copy()
```

```python
essential_nutrient_group= essential_nutrient_group.dropna().reset_index(drop=True)
```

```python
essential_nutrient_group= essential_nutrient_group.rename(columns={1:"group_name"})
```

```python
essential_nutrient_group['group_name'] = essential_nutrient_group['group_name'].replace(
    {'Carbohydrate': 'Others', 'Cholesterol': 'Others','Fiber': 'Others','Kilocalories': 'Others',
    'Protein': 'Others','Sugar Total': 'Others','Water': 'Others'})
```

```python
essential_nutrient_group= essential_nutrient_group.drop_duplicates().reset_index(drop=True)
```

```python
essential_nutrient_group.insert(0,'id_essengroup',range(1,1+len(essential_nutrient_group)))
essential_nutrient_group
```

## Essential Nutrients

```python
essential_nutrients= nutrients[[1,2]].copy()
```

```python
for index, row_value in enumerate(essential_nutrients[2].isnull()):
    if row_value:
         essential_nutrients.at[index, 2] = essential_nutrients.at[index, 1]
```

```python
essential_nutrients= essential_nutrients.dropna()
essential_nutrients.iloc[0:7,0]="Others"
essential_nutrients= essential_nutrients.rename(columns={1:"group_name",2:"nutrient_name"})
essential_nutrients.insert(0,'id_nutrient',range(1,1+len(essential_nutrients)))
essential_nutrients=essential_nutrient_group.merge(essential_nutrients, on="group_name",how='left')
essential_nutrients= essential_nutrients.dropna()
essential_nutrients=essential_nutrients.drop(['group_name'], axis=1)
essential_nutrients= essential_nutrients.reindex(columns=['id_nutrient','id_essengroup','nutrient_name'])
essential_nutrients.loc[[10],"id_essengroup"]=1
essential_nutrients
```

# Food Category

```python
#Create food category table
food_category= data[["Category"]].copy()
food_category= food_category.drop_duplicates().reset_index(drop=True)
food_category=food_category.rename(columns={"Category":"desc_category"})
food_category.insert(0,'id_category',range(1,1+len(food_category)))
food_category
```

# Food Catalog

```python
food_catalog= food_analysis_df[["Description"]].copy()
food_catalog.insert(0,'id_food',range(1,1+len(food_catalog)))
food_catalog=food_catalog.rename(columns={"Description":"food_desc"})
food_catalog
```

# Food tbl

```python
food= food_analysis_df[["Category","Description"]].copy()
column=food_analysis_df["Data.Carbohydrate"].copy()
food["nutrient_amount"]=column
food["nutrient"]= "Data.Carbohydrate"
```

```python
iterate=food_analysis_df.columns[3:]

food_tbl=pd.DataFrame()

for column_name in iterate:
    food= food_analysis_df[["Category","Description"]].copy()
    column=food_analysis_df[column_name].copy()
    food["nutrient_amount"]=column
    food["nutrient"]= column_name
    food_tbl=food_tbl.append(food, ignore_index=True)
    
food_tbl
```

```python
# Unifying measurement units mcg to mg 
for x in range(len(food_tbl)):
    if food_tbl.iloc[x,3] == "Data.Vitamins.Vitamin K" or food_tbl.iloc[x,3] == "Data.Vitamins.Vitamin B12" or food_tbl.iloc[x,3] == "Data.Vitamins.Vitamin A - RAE" :
        calories = food_tbl.iloc[x,2]
        food_tbl.iloc[x,2] = calories/1000

food_tbl
```

```python
food_tbl=food_tbl.rename(columns={"Category":"desc_category","Nutrient Data Bank Number":"id_food","Description":"food_desc"})
food_tbl=food_tbl.merge(food_category, on="desc_category",how='left')
food_tbl[['Data','Category','nutrient_name']]=food_tbl['nutrient'].str.split('.',expand=True)
food_tbl
```

```python
for index, row_value in enumerate(food_tbl['nutrient_name'].isnull()):
    if row_value:
         food_tbl.at[index, "nutrient_name"] = food_tbl.at[index, "Category"]
```

```python
food_tbl=food_tbl.merge(essential_nutrients, on="nutrient_name",how='left')
food_tbl = food_tbl.drop(['desc_category','nutrient','Data','Category','nutrient_name','id_essengroup'], axis=1)
food_tbl
```

```python
food_tbl=food_tbl.merge(food_catalog, on="food_desc",how='left')
food_tbl = food_tbl.drop(['food_desc'], axis=1)
food_tbl.insert(0,'id_foodtbl',range(1,1+len(food_tbl)))
food_tbl = food_tbl.reindex(columns=['id_foodtbl','id_food','id_category','id_nutrient','nutrient_amount'])
food_tbl
```

# Interesting facts

```python
#Extract HTML tables into DataFrames
html_link = "https://corgis-edu.github.io/corgis/csv/food/"
corgi=pd.read_html(html_link)

total_tables_found=len(corgi)
print(f"There's a Total of {total_tables_found} tables found on Corgi URL")
```

```python
#Create a Table to Manipulate for end Result
corgi_df=corgi[0]
corgi_df=corgi_df.drop(columns=['List of...', 'Example Value'])
# new data frame with split value columns 
new = corgi_df["Key"].str.split(".", n = 1, expand = True) 
  
# making separate first name column from new data frame 
corgi_df["Key1"]= new[0] 
  
# making separate last name column from new data frame 
corgi_df["Key 2"]= new[1] 
  
# Dropping old Name columns 
corgi_df.drop(columns =["Key"], inplace = True) 
  
# new data frame with split value columns 
new = corgi_df["Key 2"].str.split(".", n = 1, expand = True) 
  
# making separate first name column from new data frame 
corgi_df["NutrientGroup"]= new[0] 
  
# making separate last name column from new data frame 
corgi_df["Nutrient Name"]= new[1] 
  
# Dropping old Name columns 
corgi_df.drop(columns =["Key1"], inplace = True) 
corgi_df.drop(columns =["Key 2"], inplace = True) 
  
# df display 
corgi_df.head()
```

```python
# Multiple conditions can be set to narrow down or widen the filter
clean_corgi = corgi_df.loc[(corgi_df["NutrientGroup"] == "Fat") | (corgi_df["NutrientGroup"] == "Major Minerals") | (
    corgi_df["NutrientGroup"] == "Vitamins"), :]

clean_corgi = clean_corgi[['NutrientGroup', 'Nutrient Name', 'Comment']]
clean_corgi=clean_corgi.rename(columns={"Nutrient Name":"nutrient_name","Comment":"int_fact"})

clean_corgi=clean_corgi.merge(essential_nutrients, on="nutrient_name",how='left')
clean_corgi= clean_corgi.dropna()
clean_corgi = clean_corgi.drop(['NutrientGroup','nutrient_name','id_essengroup'], axis=1)
clean_corgi = clean_corgi.reindex(columns=['id_nutrient','int_fact'])


clean_corgi
```
